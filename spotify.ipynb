```python
df_tracks['dates']=df_tracks.index.get_level_values( 'release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot (years, discrete=True, aspect=2,height=5, kind="hist").set(title="Number of songs per year")
```

```python
# Group by 'years' and sum the 'duration' for each year
df_grouped = df_tracks.groupby(years)['duration'].sum().reset_index()

# Now extract the 'years' and 'duration' from the grouped dataframe
years_grouped = df_grouped[years.name]  # The name of the 'years' column after grouping
total_dr_grouped = df_grouped['duration']
```

```python
years_grouped = df_grouped[years.name]
```

```python
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)

# Plotting the grouped data
sns.barplot(x=years_grouped, y=total_dr_grouped, ax=ax, errwidth=0).set(title="Year vs Duration")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=90)
plt.show()
```

```python
# Group by 'years' and sum the 'duration' for each year
df_grouped = df_tracks.groupby(years)['duration'].sum().reset_index()

# Extract 'years' and 'duration' from the grouped dataframe
years_grouped = df_grouped[years.name]
total_dr_grouped = df_grouped['duration']
```

```python
sns.set_style(style="whitegrid")
fig_dims = (10, 5)
fig, ax = plt.subplots(figsize=fig_dims)

# Plot the grouped data
sns.lineplot(x=years_grouped, y=total_dr_grouped, ax=ax).set(title="Year vs Duration")

# Rotate the x-axis labels for better readability
plt.xticks(rotation=60)
plt.show()
```
